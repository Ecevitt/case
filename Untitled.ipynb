Hisseyi geÃ§miÅteki bir gÃ¼n satÄ±n alÄ±p sonraki gÃ¼nler satma hakkÄ±na sahipsiniz. 2.indexte 8 tl fiyattan alÄ±p 4.indexte 23 tlâden satmak yukarÄ±daki arrayde maksimum kar olan 23-8 = 15 TLâyi getirecektir.

Maksimumkar(3.Arr) = -1

YukarÄ±daki bilgiler Ä±ÅÄ±ÄÄ±nda;

Maksimumkar(Arr) fonksiyonu tanÄ±mlayÄ±nÄ±z. Maksimumkar fonksiyonu, input olarak alacaÄÄ± array iÃ§erisinde geÃ§miÅte satÄ±n alÄ±p gelecekte satabileceÄiniz en iyi al-sat iÅlemini uygulamalÄ±, output olarak da en yÃ¼ksek karÄ± dÃ¶ndÃ¼rmelidir. Åayet 3.Arraydaki gibi karlÄ± satÄ±Å mÃ¼mkÃ¼n deÄil ise sonuÃ§ -1 olarak dÃ¶ndÃ¼rÃ¼lmelidir.


```python
def Maksimumkar(arr):
    kar=-1
    for i in range(len(arr)):
        for k in range((i+1),len(arr)):
            if arr[k]-arr[i]>kar:
                kar=arr[k]-arr[i]
    if kar>=0:
        return kar
    else:
        return -1
```

Input =  ["B:-1", "A:1", "B:3", "A:5"]      Output = [âA:6â,âB:2â]

Yukardaki gibi bir dictionnaryde iÅlem yapÄ±p her key iÃ§in aÅaÄÄ±daki gibi alfabetik sÄ±ra ile bir toplam veren Gruptoplam(dict) fonksiyonu tanÄ±mlayÄ±nÄ±z.

Åayet value toplamlarÄ± 0 ise o keyâi outputtan exclude ediniz.


```python
def Gruptoplam(Input):
    dictionary={}
    for element in Input:
        key,value=element.split(":")
        if key in dictionary.keys():
            dictionary[key] += int(value)
        else:
            dictionary[key] = int(value)
            
    Output=[]
    for key in sorted(dictionary.keys()):
        if dictionary[key] != 0:
            Output.append(key + ":" +str(dictionary[key]))
    
    return Output
```
