```python
def Maksimumkar(arr):
    kar=-1
    for i in range(len(arr)):
        for k in range((i+1),len(arr)):
            if arr[k]-arr[i]>kar:
                kar=arr[k]-arr[i]
    if kar>=0:
        return kar
    else:
        return -1
```
